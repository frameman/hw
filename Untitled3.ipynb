```python
df = pd.read_csv("a.csv")
df.head(10)
```

```python
grouped_counts = df.groupby('縣市')['鄉鎮'].value_counts()

# Sort the results by '縣市' in ascending order
sorted_counts = grouped_counts.sort_index(level=0)
taipei_data = sorted_counts[sorted_counts.index.get_level_values('縣市') == '台北市']
b = sorted_counts[sorted_counts.index.get_level_values('縣市') == '台中市']
c = sorted_counts[sorted_counts.index.get_level_values('縣市') == '台南市']
d = sorted_counts[sorted_counts.index.get_level_values('縣市') == '宜蘭縣']
e = sorted_counts[sorted_counts.index.get_level_values('縣市') == '新竹市']
f = sorted_counts[sorted_counts.index.get_level_values('縣市') == '彰化縣']
h = sorted_counts[sorted_counts.index.get_level_values('縣市') == '桃園市']
ini = sorted_counts[sorted_counts.index.get_level_values('縣市') == '苗栗縣']
j = sorted_counts[sorted_counts.index.get_level_values('縣市') == '高雄市']
k = sorted_counts[sorted_counts.index.get_level_values('縣市') == '新北市']
l = sorted_counts[sorted_counts.index.get_level_values('縣市') == '雲林縣']

grouped_counts = df.groupby('縣市')['鄉鎮'].value_counts()

# Sort the results by '縣市' in ascending order
sorted_counts = grouped_counts.sort_index(level=0)

# Display the sorted value counts
print(sorted_counts)
```

```python
selected_column = df['縣市']

# Count the occurrences of each value in the column
value_counts = selected_column.value_counts()

# Convert the value_counts Series to a DataFrame
value_counts_df = value_counts.to_frame().reset_index()
value_counts_df.columns = ['Value', 'Count']

# Create a bar chart
plt.bar(value_counts_df['Value'], value_counts_df['Count'])
plt.xticks(rotation=45)
plt.xlabel('縣市')
plt.ylabel('確疹數字')
plt.title('猴豆確疹分佈')
for i, count in enumerate(value_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.show()
```
